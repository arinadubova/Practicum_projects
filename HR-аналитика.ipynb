```python
ohe_columns=[
    'dept',
    'last_year_promo',
    'last_year_violations'
]

ord_columns=[
    'workload',
    'level'
    
]

num_columns=[
    'employment_years',
    'salary',
    'supervisor_evaluation'
]
```

```python
RANDOM_STATE=42

x=train.drop(['job_satisfaction_rate'],axis=1)
y=train['job_satisfaction_rate']
 #–í–û–¢ –≠–¢–û –Ø –î–û–ë–ê–í–ò–õ–ê 
    
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=RANDOM_STATE)
```

OHE-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–æ–±–∞–≤–∏–º –≤ –ø–∞–π–ø–ª–∞–π–Ω

–ö–æ–¥–∏—Ä—É–µ–º —Å –ø–æ–º–æ—â—å—é OneHotEncoder:

- –ï—Å–ª–∏ –≤ –¥–∞–Ω–Ω—ã—Ö –µ—Å—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏, –∑–∞–ø–æ–ª–Ω–∏–º –∏—Ö —Å–∞–º—ã–º —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏–º—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ–º —Å –ø–æ–º–æ—â—å—é SimpleImputer. –ù–∞–∑–æ–≤—ë–º —ç—Ç–æ—Ç —à–∞–≥ simpleImputer_ohe.

- –ö–æ–¥–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ —Å –ø–æ–º–æ—â—å—é OneHotEncoder. –ù–∞–∑–æ–≤—ë–º —ç—Ç–æ —à–∞–≥ ohe.

–û–±—ä–µ–¥–∏–Ω–∏–º —à–∞–≥–∏ –≤ –ø–∞–π–ø–ª–∞–π–Ω –∏ –Ω–∞–∑–æ–≤—ë–º –µ–≥–æ ‚Äî ohe_pipe.

```python
# SimpleImputer + OHE

ohe_pipe= Pipeline(
    [
        (
            'simpleImputer_ohe',
             SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ), #strategy='most_frequent')
        (
            'ohe',
            OneHotEncoder(handle_unknown='ignore',sparse=False)
        )#handle_unknown='ignore'#handle_unknown='error' drop='first',
    ]
)
```

Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–æ–±–∞–≤–∏–º  –≤ –ø–∞–π–ø–ª–∞–π–Ω

–ö–æ–¥–∏—Ä—É–µ–º —Å –ø–æ–º–æ—â—å—é OrdinalEncoder:

- –ó–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ —Å–∞–º—ã–º —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏–º—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ–º —Å –ø–æ–º–æ—â—å—é SimpleImputer. –ù–∞–∑–æ–≤—ë–º —ç—Ç–æ—Ç —à–∞–≥ simpleImputer_before_ord.

- –ü—Ä–æ–≤–æ–¥–∏–º Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Å –ø–æ–º–æ—â—å—é OrdinalEncoder. –ù–∞–∑–æ–≤—ë–º —ç—Ç–æ—Ç —à–∞–≥ ord.

- –ó–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—è–≤–ª—è—é—Ç—Å—è –ø–æ—Å–ª–µ Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è. 

–ù–∞–∑–≤–∞–Ω–∏–µ —à–∞–≥–∞ ‚Äî simpleImputer_after_ord.


```python
#train['level'].unique()
```

```python
ord_pipe= Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['medium','high','low'],
                    ['junior','middle','senior'],
                ],
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

1. –ü—Ä–∏–º–µ–Ω—è–µ–º –ø–∞–π–ø–ª–∞–π–Ω ohe_pipe –∫ –¥–∞–Ω–Ω—ã–º ohe_columns, –∫–æ—Ç–æ—Ä—ã–µ —Ç—Ä–µ–±—É—é—Ç OHE-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è. –≠—Ç–æ —à–∞–≥ ohe.
 
2. –ü—Ä–∏–º–µ–Ω—è–µ–º –ø–∞–π–ø–ª–∞–π–Ω ord_pipe –∫ –¥–∞–Ω–Ω—ã–º ord_columns, –∫–æ—Ç–æ—Ä—ã–µ —Ç—Ä–µ–±—É—é—Ç Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è. –≠—Ç–æ —à–∞–≥ ord.

3. –ú–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ num_columns. –≠—Ç–æ —à–∞–≥ num.

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
) 
```

## –®–∞–≥ 5. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –æ–±—É—á–∏—Ç—å –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–µ –º–æ–¥–µ–ª–∏.

–û–¥–Ω—É –ª–∏–Ω–µ–π–Ω—É—é –º–æ–¥–µ–ª—å, –∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –≤—Ç–æ—Ä–æ–π ‚Äî –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π. –ü–æ–¥–±—Ä–∞—Ç—å –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–ª—è –æ–¥–Ω–æ–π –º–æ–¥–µ–ª–∏ —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é –æ–¥–Ω–æ–≥–æ –∏–∑ –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–æ–≤.

–ü–æ—Å–ª–µ –≤—ã–±–æ—Ä–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –µ—ë –∫–∞—á–µ—Å—Ç–≤–æ. –í—ã–±–æ—Ä —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è —Å–¥–µ–ª–∞—Ç—å –Ω–∞ –æ—Å–Ω–æ–≤–µ –Ω–æ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ ‚Äî SMAPE (–∞–Ω–≥–ª. symmetric mean absolute percentage error, ¬´—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ¬ª). 

**–ü–∞–π–ø–ª–∞–π–Ω –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö —Å–æ–∑–¥–∞–Ω.**

–ü—Ä–∏—Å–æ–µ–¥–∏–Ω–∏–º –∫ –Ω–µ–º—É —à–∞–≥ —Å –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏–µ–º ‚Äî –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å LinearRegression() .

–ò—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω –Ω–∞–∑–æ–≤—ë–º **pipe_final**. 

–û–Ω –≤–∫–ª—é—á–∞–µ—Ç —à–∞–≥–∏:

-–ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö ‚Äî —à–∞–≥ preprocessor.

-–ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏ ‚Äî —à–∞–≥ models.

```python
# –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
]
)
```

```python
pipe_final.fit(x, y) 
```

```python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': [None] + [i for i in range(2, 7)],
        'models__max_features': range(2,10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LinearRegression() 
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    }
]
```

```python
#np.abs(y_true - y_pred) —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∏—Å—Ç–∏–Ω–Ω—ã–º–∏ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
#((np.abs(y_true) + np.abs(y_pred)) / 2) —É—Å—Ä–µ–¥–Ω–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –∞–±—Å–æ–ª—é—Ç–Ω–æ –∏—Å—Ç–∏–Ω–Ω—ã–º –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º

def smape(y_true, y_pred):
    return 100/len(y_true) * np.sum(np.abs(y_true-y_pred)/((np.abs(y_true) + np.abs(y_pred))/2))
#c–æ–∑–¥–∞–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ —Å–∫–æ—Ä–æ–≤–µ—Ä–∞
score = make_scorer(smape, greater_is_better = False)
```

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring=score,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(x, y)

print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
print(f'–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏: {-randomized_search.best_score_}')
```

## –®–∞–≥ 6. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

–û–±—É—á–∏—Ç—å –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∫–∞–∫ –º–∏–Ω–∏–º—É–º —Ç—Ä–∏ –º–æ–¥–µ–ª–∏. –ö–∞–∫ –º–∏–Ω–∏–º—É–º –¥–ª—è –¥–≤—É—Ö –∏–∑ –Ω–∏—Ö –ø–æ–¥–±—Ä–∞—Ç—å –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏.
–ú–µ—Ç—Ä–∏–∫–∞ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –≤ —ç—Ç–æ–π –∑–∞–¥–∞—á–µ ‚Äî ROC-AUC. –ö—Ä–∏—Ç–µ—Ä–∏–π —É—Å–ø–µ—Ö–∞: ROC-AUC ‚â• 0.91 –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ. 

–ù–∞–ø–æ–º–Ω–∏–º: –æ—Ç–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —á–∞—Å—Ç–æ –ø–æ–º–æ–≥–∞–µ—Ç —É–ª—É—á—à–∏—Ç—å –º–µ—Ç—Ä–∏–∫—É.


```python
# –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
]
)
```

```python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 20),
        'models__max_features': range(2,20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1,3),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE,solver='liblinear',penalty='l1')],
        'models__C': range(1,10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_quit, y_quit)

#print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
#print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', randomized_search.best_score_)
```

```python
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', randomized_search.best_score_)
```

–ù–∞—à–ª–∏ –ª—É—á—à—É—é –º–æ–¥–µ–ª—å, –µ–π –æ–∫–∞–∑–∞–ª–∞—Å—å DecisionTreeClassifier. –ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ 0.9170512381880531/

```python
y_proba = randomized_search.predict_proba(X_test)[:, 1]  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ ROC-AUC
roc_auc = roc_auc_score(y_test, y_proba)
print(f'–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {roc_auc}')
```
